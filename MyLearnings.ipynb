**Filter Function**

```python
def even(n):   # basic Example to understand Filter function
  if n%2 == 0:
    return True

l = [1,2,3,4,5,6]

list(filter(even,l)) # filter will take func as argument
```

```python
def prime(n):
  if n<2:
    return False
  else:
    for i in range(2,n):
      if n%i == 0:
        return False
  return True

l  = [1,2,3,4,5,6,7,8,9]

list(filter(prime,l))
```

```python
# Filter funtion with lambda
l  = [1,2,3,4,5,6,7,8,9]
list(filter(lambda x: x>5 and x%2 == 0,l))
```

```python
#Using Filter funtion on list of dictionaries

mylstdicts = [
    {'name':'Arjun','Age':24},
    {'name':'Krish','Age':31},
    {'name':'Aditya','Age':23},
]

list(filter(lambda x:x['Age']>25,mylstdicts))
```

```python
def max_consecutive_difference(lst):
    max_diff = 0
    for i in range(len(lst) - 1):
        current_diff = abs(lst[i] - lst[i+1])
        if current_diff > max_diff:
            max_diff = current_diff
    return max_diff

print(max_consecutive_difference([1,2,3,4,5]))
print(max_consecutive_difference([1,2,3,6,89]))
```

```python
## Read a text file and count number of lines, words, sand charters

def count_file_content(filename):
  with open(filename,'r') as file:
    lines = file.readlines()
    line_count = len(lines)
    word_count = sum(len(line.split()) for line in lines)
    char_count = sum(len(line) for line in lines)
    return line_count,word_count,char_count

count_file_content('example.txt')
```
