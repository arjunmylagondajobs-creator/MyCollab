##OOPS Concepts

```
class Bank():
  #Constructor
  def __init__(self,name,balance = 0):
    self.name = name
    self.balance = balance

  # instance methods
  def deposit(self,amount):
    self.balance = self.balance + amount

  def withdraw(self,amount):
    if self.balance >= amount:
      self.balance = self.balance - amount
    else:
      print("Insufficient Balance")

  def get_balance(self):
    return self.balance

# Create an object

account = Bank('Arjun',25000)
print(account.name)
print(account.balance)
```

```
print(account.balance)
```

```
account.withdraw(5000)
```

```
print(account.balance)
```

```
class Team():
    # Constructor
    def __init__(self,name,position):
        self.name = name
        self.position = position

    # Creating Methods
    def resposibilities(self,position):
        if position == 'Associate':
            print("Budding Fresher with fresh mind to take up challenges")
        elif position == 'Engineer':
            print("level 1 who can do some tasks independently")
        elif position == 'SME':
            print("Senior most level can do any task independently")

    def salaries(self,position):
        if position == "Associate":
            print(30000)
        elif position == 'Engineer':
            print(40000)
        else:
            print("1lakh")

mem1 = Team('Arjun','Associate')
mem1.resposibilities('SME')
mem1.salaries('SME')
```
